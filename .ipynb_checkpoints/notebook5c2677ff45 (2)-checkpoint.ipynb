<h1>Système de Classification du Langage des Signes</h1>

<h3>Étape 1 : Configuration de l'Environnement de Travail</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf
from keras import Sequential
from keras.layers import Conv2D,MaxPool2D,Dense,Dropout,Flatten
from keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
import string
import pickle
import os
import cv2
from tqdm import tqdm
```

<h3>Étape 2 : Chargement et Exploration des Données</h3>

data link <a href="https://www.kaggle.com/datasets/grassknoted/asl-alphabet">https://www.kaggle.com/datasets/grassknoted/asl-alphabet</a>

```python
imageSize=64
folder = "../input/asl-alphabet/asl_alphabet_train/asl_alphabet_train/"
X,y = [],[]
for folderName in os.listdir(folder):
    if not folderName.startswith('.') and folderName not in ['J', 'Z', 'space', 'del', 'nothing']:
        for image_filename in tqdm(os.listdir(folder + folderName)):
            img = cv2.imread(folder + folderName + '/' + image_filename)
            if img is not None:
                img = cv2.resize(img, (imageSize, imageSize))
                img_arr = np.asarray(img)
                X.append(img_arr)
                y.append(folderName)

X = np.asarray(X)
y = np.asarray(y)
```

```python
X.shape,y.shape
```

```python
X_flatten = X.reshape(X.shape[0], -1)
```

```python
data = pd.DataFrame(X_flatten, columns=[f'pixel{i}' for i in range(1, X_flatten.shape[1] + 1)])
data['label'] = y
```

```python
# Explorer les données du train
print(data.head())
print(data.info())
print(data.describe())
```

```python
# Distribution of labels in the dataset
label_distribution = data['label'].value_counts().sort_index()

# Plotting the distribution
plt.figure(figsize=(10, 6))
label_distribution.plot(kind='bar')
plt.title('Distribution of Labels in the Sign Language Dataset')
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

<h3>Étape 3 : Prétraitement des Données Statiques</h3>


```python
# Séparation des Étiquettes et des Pixels

images_train,images_test,labels_train,labels_test = train_test_split(X,y,test_size=.2,random_state=42)
```

<h4>Normalization</h4>

```python
# Normalisation des pixels
images_train = images_train / 255.0
images_test = images_test / 255.0
```

<h4>plot quelques images</h4>

```python
plt.imshow(images_train[0], cmap='gray')
plt.show()
```

```python
# Visualizing the first few images in the training set
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images_train[i], cmap='gray')
    plt.title(f'Label: {labels_train[i]}')
    plt.axis('off')
plt.show()

images_train.shape, images_test.shape, labels_train.shape, labels_test.shape
```

<h4>Augmentation des Données (Data Augmentation)</h4>

```python
#Augmentation des Données (Data Augmentation) pour l'ensemble d'entraînement
datagen_train = ImageDataGenerator(
    rotation_range=10,  # degrés de rotation aléatoire
    width_shift_range=0.1,  # décalage horizontal aléatoire
    height_shift_range=0.1,  # décalage vertical aléatoire
    zoom_range=0.1  # plage de zoom aléatoire
)

# Appliquer l'augmentation uniquement sur l'ensemble d'entraînement
datagen_train.fit(images_train)
```

<h4>Conversion des labels entières en labels catégorielles </h4>

```python
lb=LabelBinarizer()
labels_train=lb.fit_transform(labels_train)
labels_test=lb.transform(labels_test)

print("shape of labels test : ",labels_train.shape)
print("shape of labels test : ",labels_test.shape)
```

```python
labels_train[:5]
```

<h4>Enregistrement de l'objet LabelBinarizer avec Pickle</h4>

```python
save_dir = 'saved_objects'  # Directory name
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
file_path = os.path.join(save_dir, 'label_binarizer.pkl')
with open(file_path, 'wb') as file:
    pickle.dump(lb, file)
    print(f"LabelBinarizer saved successfully at '{file_path}'")
```

<h3>Construction du Modèle</h3>

```python
model=Sequential([
                Conv2D(64,kernel_size=(5,5),
                        strides=1,padding='same',activation='relu',input_shape=(imageSize,imageSize,3)),
                MaxPool2D(pool_size=(3,3),strides=2,padding='same'),
                Conv2D(64,kernel_size=(2,2),
                        strides=1,activation='relu',padding='same'),
                MaxPool2D((2,2),2,padding='same'),
                Conv2D(32,kernel_size=(2,2),
                        strides=1,activation='relu',padding='same'),
                MaxPool2D((2,2),2,padding='same'),

                Flatten(),
                Dense(units=512,activation='relu'),
                Dropout(rate=0.25),
                Dense(units=24,activation='softmax'),
    ])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Define EarlyStopping callback
early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

# Example usage in model training
history = model.fit(
    datagen_train.flow(images_train, labels_train, batch_size=200),
    epochs=20,
    validation_data=(images_test, labels_test),
    shuffle=1,
    callbacks=[early_stop]
)
```

<h3>Évaluation du Modèle</h3>

```python
model.evaluate(images_test,labels_test)
```

<h3>Évolution des performances pendant l'apprentissage du modèle</h3>

```python
# Assuming history contains accuracy and loss data from model training
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))  # Assuming epochs range based on actual training

plt.figure(figsize=(10, 8))

# Plotting Accuracy
plt.subplot(2, 1, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xticks(epochs_range, epochs_range)  # Set x-axis ticks to all values in epochs_range

# Plotting Loss
plt.subplot(2, 1, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xticks(epochs_range, epochs_range)  # Set x-axis ticks to all values in epochs_range

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
y_pred=model.predict(images_test)
y_pred[0]
```

```python
threshold = 0.5
# Convert probabilities to binary labels using the threshold
binary_labels = np.where(y_pred >= threshold, 1, 0)
binary_labels[0]
```

```python
predicted_labels=lb.inverse_transform(binary_labels)
y_true=lb.inverse_transform(labels_test)
```

<h3>Marice de confusion</h3>

```python
#calculate confusion matrix
conf_matrix = confusion_matrix(y_true, predicted_labels)

class_labels = [c for c in string.ascii_uppercase if c != 'J' and c!='Z']
plt.figure(figsize=(15, 12))

# Create a heatmap using Seaborn
sns.set(font_scale=1.2)  # Adjust font size if necessary
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_labels, yticklabels=class_labels)

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix (24x24)')
plt.show()
```
