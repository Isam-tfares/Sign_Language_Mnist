```python
model3 = tf.keras.models.load_model("./model.h5")
```

```python
import pickle
# Loading LabelBinarizer (lb) from the saved file
with open('./saved_objects/label_binarizer.pkl', 'rb') as file:
    loaded_lb = pickle.load(file)
def get_label(y):
    return loaded_lb.inverse_transform(y)[0][0]
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
path_folder="./data/asl_alphabet_test/asl_alphabet_test/"
plt.figure(1,figsize=(16,16))
i=1
for img_name in os.listdir(path_folder):
    lettre=img_name.split('.')[0].split('_')[0]
    if lettre == "Z" or lettre=="J" or lettre=="nothing" or lettre=="space":
        continue
    img=cv2.imread(os.path.join(path_folder,img_name))
    x=cv2.resize(img,(64,64))
    x=x/255.0
    y=model3.predict(x.reshape((1,64,64,3)))[0]
    plt.subplot(6,4,i)
    i=i+1
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    yp=np.argmax(y)
    y=np.zeros((1,24))
    y[0][yp]=1
    plt.title(lettre +" predicted "+get_label(y))
    plt.xticks([])
    plt.yticks([])
    
```
