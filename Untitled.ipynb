```python
import cv2
import mediapipe as mp
from predict import predict_image
# from tensorflow import keras
```

```python
# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils  # Add this line
hands = mp_hands.Hands()

# Function to detect and draw hand landmarks on the frame
def detect_hand(frame):
    # Convert the BGR image to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Process the frame with MediaPipe Hands
    results = hands.process(rgb_frame)

    # Check if hands are detected
    if results.multi_hand_landmarks:
        # Draw landmarks on the frame
        for hand_landmarks in results.multi_hand_landmarks:
#             mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

            # Extract bounding box coordinates
            bboxC = [(int(lm.x * frame.shape[1]), int(lm.y * frame.shape[0])) for lm in hand_landmarks.landmark]
            x_min, y_min = min(bboxC)[0], min(bboxC, key=lambda x: x[1])[1]
            x_max, y_max = max(bboxC)[0], max(bboxC, key=lambda x: x[1])[1]

            # Draw a green rectangle around the hand
            cv2.rectangle(frame, (x_min - 20, y_min - 20), (x_max + 20, y_max + 20), (0, 255, 0), 2)
            hand=frame[y_min - 20:y_max + 20,x_min - 20:x_max + 20]
    
    return frame,hand


cap = cv2.VideoCapture("./ASL video.mp4")
i=0
old_letter=None
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detect and draw hand landmarks
    frame_with_landmarks,hand = detect_hand(frame)
    if i%20==0 or old_letter==None:
        old_letter=predict_image(hand)['prediction']
        
    # writte letter
    cv2.putText(frame, f'Prediction: {old_letter}', (200, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0,0), 2)

    # Display the frame
    cv2.imshow('Hand Detection', frame_with_landmarks)
    i+=1

    # Break the loop with 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cap.release()
cv2.destroyAllWindows()
```
